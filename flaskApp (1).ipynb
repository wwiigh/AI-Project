```python
from flask import Flask, request, jsonify
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer



from pyngrok import ngrok

ngrok.set_auth_token('2geFIz7UurOg31GG5gPrbr4pUBp_ZtPYVjGrvVGkVaLpzRfd')
# Set up ngrok with a specific port
public_url = ngrok.connect(5000)
print("ngrok tunnel \"{}\"".format(public_url))

from flask_ngrok import run_with_ngrok
from flask import Flask, request, jsonify
from flask_cors import CORS

app = Flask(__name__)
CORS(app)


# Define a global device variable
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Function definitions
def load_model(model_path):
    model = GPT2LMHeadModel.from_pretrained('gpt2')
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device)
    model.eval()
    return model

models = {
    "vegetarian": load_model('/content/drive/MyDrive/Colab_Notebooks/Intro_to_AI/Final_project/models/vegan.pt'),
    "lacto-ovo vegetarian": load_model('/content/drive/MyDrive/Colab_Notebooks/Intro_to_AI/Final_project/models/lacto_ovo.pt'),
    "none": load_model('/content/drive/MyDrive/Colab_Notebooks/Intro_to_AI/Final_project/models/all.pt')
}

# Set the tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token


@app.route("/generate-recipe", methods=['POST'])
def generate_recipe():
    data = request.get_json()
    title = data['title']
    ingredients = data['ingredients']
    diet = data['diet']

    raw_text = 'Use: '+ ingredients + '. Title: ' + title
    prompt = tokenizer.encode(raw_text, return_tensors="pt").to(device)

    # Create attention mask (1 for actual tokens, 0 for padding)
    attention_mask = torch.ones(prompt.shape, dtype=torch.long, device=device)
    attention_mask[prompt == tokenizer.pad_token_id] = 0

    model = models.get(diet, models['none'])

    # Generate output with the attention mask
    output = model.generate(
        prompt,
        attention_mask=attention_mask,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        max_length=300,
        top_k=10,
        temperature=0.8
    )

    output_text = tokenizer.decode(output[0], skip_special_tokens=True)
    print("output_text: ")
    print(output_text)

    import re

    # Define regex patterns
    title_pattern = r"Title:\s*(.+?)(?:\n|$)"
    ingredients_pattern = r"Ingredients:\s*(.+?)(?:\. Title:.*|\n|$)"
    directions_pattern = r"Directions:\s*(.+)"

    # Search for title
    title_search = re.search(title_pattern, output_text, re.IGNORECASE | re.DOTALL)
    title = title_search.group(1).strip() if title_search else "No title found"

    # Search for ingredients
    ingredients_search = re.findall(ingredients_pattern, output_text, re.IGNORECASE | re.DOTALL)
    ingredients = [ingredient.strip() for ingredient in ingredients_search]

    # Search for directions
    directions_search = re.search(directions_pattern, output_text, re.IGNORECASE | re.DOTALL)
    directions = directions_search.group(1).strip() if directions_search else "No directions found"

    # Print extracted data
    print("extracted data: ")
    print("Title:", title)
    print("Ingredients:", ingredients)
    print("Directions:", directions)

    return jsonify({
        'recipe': {
            'title': title,
            'ingredients': ingredients,
            'directions': directions
        }
    })
if __name__ == "__main__":
    app.run()
```
